```python
import cv2
import pandas as pd
import tqdm
from tqdm import trange
import numpy as np
import os
from google.colab.patches import cv2_imshow
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("/content/drive/MyDrive/data_project/1.jpg")
```

```python
def display_img(cols, rows, **img):
  fig = plt.figure(figsize=(25,14))
  for i in range(cols*rows):
    title = img["title"][i]
    image = img["img"][i].copy()
    fig.add_subplot(rows, cols, i+1)
    plt.title(title, size = 19)
    try:
      image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    except:
      image = img["img"][i].copy()
    plt.imshow(image, interpolation= "bicubic")
    plt.axis("off")
```

```python
def Btgraham(img, scale):

    a=cv2.addWeighted(img, 4, cv2.GaussianBlur(img, (0,0) ,scale/30) , -4 ,128)

    return a.astype("uint8")
```

```python

bt_img = Btgraham(img,300)
display_img(2,1 ,img = [img, bt_img], title = ["image", "Btgraham"])
cv2.imwrite("/content/drive/MyDrive/data_project/processed_1.jpg",bt_img)
```
